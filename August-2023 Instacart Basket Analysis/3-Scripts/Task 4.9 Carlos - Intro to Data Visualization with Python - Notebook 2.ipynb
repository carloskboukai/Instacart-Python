Making a scatterplot. I am using a sample of the data. I was not sure of the first scatterplott results. However, using just a sample of the data (30%), I achieved the same results.

While the scatterplot shows that most people of all ages are situated below the 200000 mark for income, it also shows that from the age of 40 there is a very significant part of the customers that earn up to 300000 (which is 50% more income). Also, it is from the age of 40 that one can see that there are customers that earn up to 600000 (almost 200% more). Below 40, there are no entries after the mark of 400000. Clearly, it seems that - potentially, at least - that income may increase with age. It would be logical since, individuals over 40 should be more established both financially and with families.
