02. Bar chart => family status profile

```python
# Adding colour to the plot
barsub = subfinal['family_situation'].value_counts().plot.barh(color = ['blue', 'darkorange', 'cornflowerblue', 'navajowhite', 'cyan'], title = 'Family Status Profile')
```

04. Line chart - trend of purchase day of week

```python
line = sns.lineplot(data = subfinal, x = 'orders_day_of_week',y = 'family_situation')
```

The graphic shows that (in general) families tend to order more on Saturday (in preparation for the weekend) and also on midweek (probably for restocking). While there are 5 categories of 'family' which includes single men and women, more than 80% of customers have families, which indicates why the trend looks as it does.

```python
# Exporting and saving it to Folder
line.figure.savefig(os.path.join(path, '4-Analysis','Visualisations', 'line_family_dow.png'))
```

05. Line chart - trend of purchase hour of day

```python
line = sns.lineplot(data = subfinal, x ='order_hour_of_day',  y = 'family_situation').set(title = 'Family hourly buying trend')
```

The line indicates that most orders happen throughout the day between 9:30am to 8:30pm. The spike during non sociable hours is likely to be from single men and women which are not bound by normal family 'time constraints'. The scatterplot below seems to confirm this.

```python
# Exporting and saving it to Folder
line.figure.savefig(os.path.join(path, '4-Analysis','Visualisations', 'line_family_hourofday.png'))
```

```python
# Attempting a Scatterpolot => Hour of day trend
scat = sns.scatterplot(x = 'family_situation', y = 'order_hour_of_day',data = subfinal)
```

```python
# Exporting and saving it to Folder
scat.figure.savefig(os.path.join(path, '4-Analysis','Visualisations', 'scat_family_hourofday.png'))
```

06. Bar chart => figures of department list

```python
# Adding colour to the plot
barsub3 = subfinal['department'].value_counts().plot.barh(color = ['darkgreen', 'red', 'green', 'tomato', 'darkolivegreen', 'lightcoral', 'limegreen', 'salmon', 'mediumseagreen', 'orangered', 'mediumspringgreen', 'lightsalmon', 'greenyellow', 'coral','lightgreen'], title = 'Orders from Departments')
```

It is clearly that most orders come from regular/restockable food items such as produce, dairy and eggs. These are items that are commonly necessary for families with dependants, and it is consistent with the proportion of families (not single adults) of Instacart's customers' population.

```python
# Exporting and saving it to Folder
barsub3.figure.savefig(os.path.join(path, '4-Analysis','Visualisations', 'department_figures.png'))
```
